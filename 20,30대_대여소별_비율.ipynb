```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np  
```

```python
from google.colab import drive
drive.mount('/content/drive')

bike = pd.read_excel("/content/drive/My Drive/Colab Notebooks/ITB/용산구16.xlsx")
```

```python
bike['성별'] = np.where(bike['성별'] == 'm', 'M', bike['성별'])
bike['성별'] = np.where(bike['성별'] == 'f', 'F', bike['성별'])
```

```python
bike['운동량'] = np.where(bike['운동량'] == r'\N', np.nan, bike['운동량'])
```

```python
bike= bike.dropna()
```

```python
bike['월'] = np.where(bike['대여일자'] == '2022-01-01', '1월',
                     np.where(bike['대여일자'] == '2022-02-01', '2월',
                               np.where(bike['대여일자'] == '2022-03-01', '3월',
                                         np.where(bike['대여일자'] == '2022-04-01', '4월',
                                                   np.where(bike['대여일자'] == '2022-05-01', '5월', '6월')))))
```

```python
bike['역주변'] = bike['대여소명'].str.contains('역')
```

```python
bike['학교주변'] = bike['대여소명'].str.contains('학교|여대')
```

```python
bike
```

```python
bike_n = bike.groupby ('대여소명', as_index = False)\
 .agg (n = ('이용건수','sum')).sort_values('n', ascending=False)
bike_n
```

```python
bike_n['역주변'] = bike['대여소명'].str.contains('역')
bike_n
```

#2030건수, 비율

```python
#20대 대여소 이용건수
bike_n20 = bike.query ('연령대코드 == "20대"').groupby ('대여소명', as_index = False)\
 .agg (n_20 = ('이용건수','sum')).sort_values('n_20', ascending=False)
bike_n20
```

```python
#30대 대여소 이용건수 
bike_n30 =bike.query ('연령대코드 == "30대"').groupby ('대여소명', as_index = False)\
 .agg (n_30 = ('이용건수','sum')).sort_values('n_30', ascending=False)
bike_n30
```

```python
#2030대 대여소 이용건수
bike_n2030 = bike.query ('연령대코드 == "30대" or 연령대코드 == "20대"').groupby ('대여소명', as_index = False)\
 .agg (n_2030 = ('이용건수','sum')).sort_values('n_2030', ascending=False)
bike_n2030
```

```python
#총 이용건수 : 2030대 이용건수
bike_nn = pd.merge(bike_n,bike_n2030)
```

```python
bike_nn['비율'] = bike_nn['n_2030']/bike_nn['n']
bike_nn
```

```python
bike_nn.sort_values('비율', ascending=False)
```

```python
#총이용건수 중 2030 이용건수 비율_상위
bike_nn.sort_values('비율', ascending=False).head(20)
```

```python
#총이용건수 중 2030 이용건수 비율_하위
bike_nn.sort_values('비율').head(10)
```

```python
#전체 이용건수 총합 중 역주변 이용건수
bike_n.groupby ('역주변', as_index = False)\
 .agg (n = ('n','sum')).sort_values('n', ascending=False)
```

```python
#역주변 이용건수 
bike_sn =bike.query ('역주변 == True').groupby ('대여소명', as_index = False)\
 .agg (n = ('이용건수','sum')).sort_values('n', ascending=False)
```

```python
#역주변 2030대 이용건수 
bike_sn2030 = bike.query ('역주변 == True and (연령대코드 == "20대" or 연령대코드 =="30대")' ).groupby ('대여소명', as_index = False)\
 .agg (n_2030 = ('이용건수','sum')).sort_values('n_2030', ascending=False)
```

```python
bike_snn = pd.merge(bike_sn,bike_sn2030)
bike_snn['비율'] = bike_snn['n_2030']/bike_nn['n']
bike_snn
```

```python
#역주변 총이용건수 중 2030 이용건수 비율_상위
bike_snn.sort_values('비율', ascending=False).head(5)
```

```python
#학교주변 총이용건수 
bike.query ('학교주변 == True').groupby ('대여소명', as_index = False)\
 .agg (n = ('이용건수','sum')).sort_values('n', ascending=False)
```
